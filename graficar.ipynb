```python
import pandas as pd
import numpy as np

# Seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 1000

# Generating Order IDs
order_ids = range(1, num_samples + 1)

# Generating Product Categories
product_categories = np.random.choice(
    ["Electronics", "Clothing", "Home & Kitchen", "Sports"], num_samples
)

# Generating Order Dates
order_dates = pd.date_range(start="2023-01-01", periods=num_samples, freq="H")

# Generating Customer Ages
customer_ages = np.random.randint(18, 70, num_samples)

# Generating Countries
countries = np.random.choice(
    ["USA", "Canada", "Germany", "France", "Australia", "Spain"], num_samples
)

# Generating Order Amounts
order_amounts = np.round(np.random.uniform(20, 1000, num_samples), 2)

# Generating Payment Methods
payment_methods = np.random.choice(
    ["Credit Card", "PayPal", "Bank Transfer"], num_samples
)

# Generating Shipping Times
shipping_times = np.random.randint(1, 15, num_samples)

# Creating the DataFrame
df = pd.DataFrame({
    "Order ID": order_ids,
    "Product Category": product_categories,
    "Order Date": order_dates,
    "Customer Age": customer_ages,
    "Country": countries,
    "Order Amount": order_amounts,
    "Payment Method": payment_methods,
    "Shipping Time": shipping_times
})


```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
orders_by_date = df.groupby(df['Order Date'].dt.date)['Order Amount'].sum()

plt.figure(figsize=(10,6))
plt.plot(orders_by_date.index, orders_by_date.values)
plt.title('vaya mierda')
plt.xlabel('pa tu madre cabron')
plt.tight_layout()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data=df, x='Customer Age', y='Order Amount', alpha=0.56)
plt.title('la carrasca')
plt.ylabel('carrasca tu mamon')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Customer Age'], bins=30, color='green',  edgecolor='black')
```

```python
total_by_category = df.groupby('Product Category')['Order Amount'].sum()

plt.figure(figsize=(10, 6))
plt.bar(total_by_category.index, total_by_category.values)
```

```python
import seaborn as sns# Crear el violin plot con Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x="Payment Method", y="Order Amount", data=df, palette="Set3")
plt.title("Distribución del Monto de los Pedidos por Método de Pago (Violin Plot)")
plt.xlabel("Método de Pago")
plt.ylabel("Monto del Pedido")
plt.grid(True)
plt.show()
```

```python
# Pairplot con Seaborn para explorar relaciones entre variables numéricas
sns.pairplot(df[["Customer Age", "Order Amount", "Shipping Time"]], diag_kind='hist', palette="Set2")
plt.suptitle("Pairplot de Edad del Cliente, Monto del Pedido y Tiempo de Envío", y=1.02)
plt.show()
```

```python
# Crear el swarmplot con Seaborn
plt.figure(figsize=(10, 6))
sns.swarmplot(x="Payment Method", y="Order Amount", data=df, palette="Set2")
plt.title("Swarmplot del Monto de los Pedidos por Método de Pago")
plt.xlabel("Método de Pago")
plt.ylabel("Monto del Pedido")
plt.grid(True)
plt.show()
```
