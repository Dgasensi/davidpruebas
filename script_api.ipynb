```python
# Function to generate the prompt based on the final structure
def generate_six_conversations_prompt(brand):
    prompt = (f"Assume the role of a chatbot assistant for {brand} that has been fine-tuned specifically "
              f"to assist the customers of {brand}. Your main mission is to help {brand}'s customers with their "
              f"inquiries and provide excellent customer service.\n\n"
              f"Generate 6 simulated conversations between a customer and the assistant from {brand}. "
              f"The conversations should cover interactions relevant to {brand}'s products or services.\n\n"
              f"- Out of these, 2 conversations should have a positive sentiment, 2 should have a negative sentiment, and 2 should be neutral.\n"
              f"- The conversations should have between 4 and 12 turns (messages exchanged between the customer and the assistant).\n"
              f"- Return the result in the following JSON format:\n\n"
              f"chat_data_example = [\n"
              f"  {{\n"
              f"    \"conversation_id\": \"12345\",\n"
              f"    \"model\": \"gpt-4\",\n"
              f"    \"system_prompt\": \"You are a helpful assistant.\",\n"
              f"    \"messages\": [\n"
              f"      {{\"sender\": \"user\", \"timestamp\": \"2024-07-28T10:00:00Z\", \"content\": \"Hello, how are you?\"}},\n"
              f"      {{\"sender\": \"assistant\", \"timestamp\": \"2024-07-28T10:00:10Z\", \"content\": \"I'm good, thank you! How can I assist you today?\"}},\n"
              f"      {{\"sender\": \"user\", \"timestamp\": \"2024-07-28T10:01:00Z\", \"content\": \"Can you help me with a Python question?\"}},\n"
              f"      {{\"sender\": \"assistant\", \"timestamp\": \"2024-07-28T10:01:10Z\", \"content\": \"Sure, what do you need help with?\"}}\n"
              f"    ]\n"
              f"  }}\n"
              f"]\n"
              f"- The timestamps and conversation_id should be unique for each message.\n")
    
    return prompt

# Function to generate conversations using the OpenAI API
def generate_six_conversations(brand):
    prompt = generate_six_conversations_prompt(brand)
    
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a helpful virtual assistant."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=3000  # Adjust based on expected length of the conversations
    )
    
    return response['choices'][0]['message']['content']

# Main function to save all conversations to a single JSON file at the end
def save_all_conversations_to_json(all_conversations, filename="all_conversations.json"):
    with open(filename, "w") as outfile:
        json.dump(all_conversations, outfile, indent=4)

# Main logic to generate conversations for each brand and accumulate them
all_conversations = []

brands_and_sectors = [(brand, sector) for sector in sectors for brand in sectors[sector]]

# Loop over the brands and generate six conversations for each
for brand, sector in brands_and_sectors:
    conversations = generate_six_conversations(brand)
    
    try:
        # Parse the response and add to the list
        conversation_data = json.loads(conversations)
        # Append to the main list of conversations
        all_conversations.extend(conversation_data)
    
    except json.JSONDecodeError:
        print(f"Error decoding JSON for brand {brand}. Skipping this conversation.")

# Save all conversations to a single JSON file
save_all_conversations_to_json(all_conversations)
```
