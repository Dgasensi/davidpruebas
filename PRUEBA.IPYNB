```python
import pandas as pd 
import matplotlib
df = pd.read_csv('jobs_in_data.csv')
pd.set_option('display.max_columns' ,None)
df.head(5)
```

```python
company_loc = pd.unique(df['company_location'])
salario_max_habitar = pd.DataFrame(df.groupby('company_location')['salary'].mean())
df_spain = df[df['employee_residence'] == 'Spain']
df_spain.sort_values(by='salary')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#hallar sueldo medio

#GRAFICAR MEDIA SUELDO POR  CATEGORIA
df_spain = df[df['employee_residence'] == 'Spain']
df_ger = df[df['employee_residence'] == 'Germany']
sueldo_medio = df_spain.groupby('experience_level')['salary'].mean()
df_usa = df[df['employee_residence'] == 'United States']


plt.figure(figsize=(15, 6))
plt.bar(df_spain['experience_level'], df_spain['salary'],width=0.15, color='blue')
plt.bar(df_ger['experience_level'], df_ger['salary'],width=0.35, align='edge', color='red')
plt.bar(df_usa['experience_level'], df_usa['salary'],width=0.15, align=('edge'), color='green')


plt.title('sueldos españa')
plt.show()
```
